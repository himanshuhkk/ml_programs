# Matplotlib

```python
from matplotlib import pyplot as plt
x = np.arange(1,11)
y = 2*x
y1 = 5*x
plt.plot(x,y , color = 'r' , linestyle = ':' , linewidth = 5)
plt.plot(x,y1, color = 'b'  , linewidth = 5)
plt.title("plotting")
plt.xlabel('x')
plt.ylabel('y')
plt.show()
plt.subplot(1,2,1)
plt.plot(x,y,color = 'r')
plt.subplot(1,2,2)
plt.plot(x,y1 , color = 'g')
plt.show()
# barplots
bar = {'honey':10 , 'ankit':102 , 'dharmu':60}
names = list(bar.keys())
values = list(bar.values())
plt.bar(names , values)
plt.show()
plt.scatter(x,y,color='r')
plt.show()
plt.scatter(x,y1,marker = '*' , color='g' , s=100)
plt.show()
plt.hist(x , color='g' , bins = 10)
plt.show()
plt.boxplot([x,y,y1])
plt.show()
plt.violinplot([x,y,y1])
plt.show()
companies = ['jio' , 'tata' , 'aditya birla' , 'wipro']
customers = [10000 , 12000 , 9000 , 5000]
plt.pie( customers , labels = companies , autopct = '%0.2f%%' , colors = ['cyan' , 'magenta' , 'purple' , 'yellow'] )
plt.show()
plt.pie( customers , labels = companies , radius = 2 , autopct = '%0.2f%%' , colors = ['cyan' , 'magenta' , 'purple' , 'yellow'] )
plt.pie([1] , colors = 'w' , radius = 1)
plt.show()
```
