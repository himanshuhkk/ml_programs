```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
data_frame = pd.DataFrame(breast_cancer_dataset.data , columns=breast_cancer_dataset.feature_names)
```

```python
data_frame['label'] = breast_cancer_dataset.target
```

```python
x = data_frame.drop(columns = 'label' , axis = 1)
y = data_frame['label']
```

```python
x.head()
```

```python
y.head()
```

```python
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size = 0.2 , random_state=2)
```

```python
# standard scalar as values of columns are in different range
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_std = scaler.fit_transform(x_train)
x_test_std = scaler.transform(x_test)  # donot fit testing data only fit train data
```

```python
import tensorflow as tf
tf.random.set_seed(3)  # As weight and parameter in neural networks keeps on changing so accuracy may change on each run so to avoid this , we used seed to get same accuracy . 
from tensorflow import keras
```

```python
# setting up the layer for neural network
model = keras.Sequential([
    keras.layers.Flatten(input_shape = (30 , )) ,         # flatten means to convert training or testing data into single dimensional array . so each feature will go to one neuron . 
    # so number of neurons in input layer = no. of features in dataset . (for numerical data , different for image classification)
    #input_shape is no. of features or neurons . 
    keras.layers.Dense(20 , activation='relu')  ,  
    #dense means all neurons of hidden layer is connected to all neurons in previous layer
    # each neuron has its activation function . 
    # In hidden layer we have given 20 neurons . 
    keras.layers.Dense(2 , activation='sigmoid')
    #output layer contains 2 neurons for output (firing of neuron means which neuron is giving output or is working)
])
```

```python
# compiling neural network
model.compile(optimizer='adam' , 
             loss = 'sparse_categorical_crossentropy' , #  for categorical output as 0 , 1 , 2
              # one hot encoding if output is like 100 , 010 , 001 means if u have more outputs say 1 to 20 go for this . 
             metrics=['accuracy'])
```

```python
# training
history = model.fit(x_train_std , y_train , validation_split=0.1 , epochs=8)   # epochs means how many times model has to go through the data
```

```python
# visualising accuracy and loss
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['training data' , 'validation data'] , loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['training data' , 'validation data'] , loc = 'upper right')
```

```python
# get accuracy of model on test data
loss , accuracy = model.evaluate(x_test_std , y_test)
```
